# DeepSphere Experiments: A "Winning Model"?

One of our models with a fully convolutional architecture and five layers seemed to successfully reach 0.02 L1 loss on the validation set. If these constraints hold for noisy galaxy count maps across the board, we have a winner!

But how accurate is this network exactly? Does it hold scrutiny to further testing on all training quartiles, the full test set and varying noise/smoothing scales?

## Re-creating Model

```python
params = dict()
params['dir_name'] = EXP_NAME

# Types of layers.
params['conv'] = 'chebyshev5'  # Graph convolution: chebyshev5 or monomials.
params['pool'] = 'max'  # Pooling: max or average.
params['activation'] = 'relu'  # Non-linearity: relu, elu, leaky_relu, softmax, tanh, etc.
params['statistics'] = 'mean'  # Statistics (for invariance): None, mean, var, meanvar, hist.

# Architecture.
params['F'] = [16, 32, 64, 64, 1]  # Graph convolutional layers: number of feature maps.
params['K'] = [5] * 5  # Polynomial orders.
params['batch_norm'] = [True, True, True, True, False]  # Batch normalization.
params['M'] = []  # Fully connected layers: output dimensionalities.
params['input_channel'] = 1  # Two channels (spherical maps) per sample.

# Pooling.
nsides = [NSIDE, NSIDE//2, NSIDE//4, NSIDE//8, NSIDE//16, NSIDE//16]
params['nsides'] = nsides
params['indexes'] = utils.nside2indexes(nsides, ORDER)
#     params['batch_norm_full'] = []

# Regularization (to prevent over-fitting).
params['regularization'] = 0  # Amount of L2 regularization over the weights (will be divided by the number of weights).
params['dropout'] = 1  # Percentage of neurons to keep.

# Training.
params['num_epochs'] = 20  # Number of passes through the training data.
params['batch_size'] = 16 * ORDER**2  # Constant quantity of information (#pixels) per step (invariant to sample size).

# Optimization: learning rate schedule and optimizer.
params['scheduler'] = lambda step: tf.train.exponential_decay(1.99e-4, step, decay_steps=1, decay_rate=0.999)
params['optimizer'] = lambda lr: tf.train.AdamOptimizer(lr, beta1=0.9, beta2=0.999, epsilon=1e-8)
params['loss'] = 'l1'  # Regression loss.

# Number of model evaluations during training (influence training time).
params['eval_frequency'] = 15

print('#sides: {}'.format(nsides))
print('#pixels: {}'.format([(nside//ORDER)**2 for nside in nsides]))
# Number of pixels on the full sphere: 12 * nsides**2.

print('#samples per batch: {}'.format(params['batch_size']))
print('=> #pixels per batch (input): {:,}'.format(params['batch_size']*(NSIDE//ORDER)**2))
print('=> #pixels for training (input): {:,}'.format(params['num_epochs']*422*(NSIDE//ORDER)**2))
```

```python
model = models.deepsphere(**params)
```

```python
def plot_residuals(ax, true, pred, lab="predictions", 
                   ti='"Winning Model" prediction of parameter {}'.format("$\sigma_8$"), fs=14):
    ax.plot(true, pred, 'bo', label=lab)
    ax.plot([0.5, 1.2], [0.5, 1.2], 'r-', linewidth=6, label='ground truth')
    ax.set_xlabel('Ground truth value of $\sigma_8$', fontsize=fs)
    ax.set_ylabel('Predicted value of $\sigma_8$', fontsize=fs)
    ax.set_title(ti, fontsize=fs)
    ax.legend()
```

```python
def l1_loss(true, pred):
    return np.mean(np.abs(true-pred))
```

```python
l1q1 = l1_loss(y_train1[:,0], pred_train1[:,0])
l1q2 = l1_loss(y_train2[:,0], pred_train2[:,0])
l1q3 = l1_loss(y_train3[:,0], pred_train3[:,0])
l1q4 = l1_loss(y_train4[:,0], pred_train4[:,0])
l1test = l1_loss(y_test[:,0], pred_test[:,0])
```

```python
def conf(labels, preds):
    cosmos = np.sort(np.unique(labels))    
    n = len(cosmos)
    preds_by_cosmo = np.zeros((n, 48, 1))
    for i in range(n):
        bool_cosmo = labels[:,0] == cosmos[i]
        preds_by_cosmo[i,:] = preds[bool_cosmo]
    preds_by_cosmo = np.reshape(preds_by_cosmo, (n, 48))
    std_by_cosmo = np.std(preds_by_cosmo, axis=1)
    return np.mean(std_by_cosmo)
```

```python
sq1 = conf(y_train1, pred_train1)
sq2 = conf(y_train2, pred_train2)
sq3 = conf(y_train3, pred_train3)
sq4 = conf(y_train4, pred_train4)
stest = conf(y_test, pred_test)
```

```python
print("WINNING MODEL CONFIDENCE INTERVALS:")
sets = ["Train Q1", "Train Q2", "Train Q3", "Train Q4", "Test"]
stds = [sq1, sq2, sq3, sq4, stest]
for s, l in zip(sets, stds):
    print("Mean L1 Sigma on {0}: {1}".format(s, l))
print("Combined Training Sigma: {}".format(np.mean(stds[:-1])))
print("Combines Training/Testing Sigma: {}".format(np.mean(stds)))
```

```python
print("WINNING MODEL L1 LOSSES:")
sets = ["Train Q1", "Train Q2", "Train Q3", "Train Q4", "Test"]
losses = [l1q1, l1q2, l1q3, l1q4, l1test]
for s, l in zip(sets, losses):
    print("Mean L1 Loss on {0}: {1}".format(s, l))
print("Combined Training Loss: {}".format(np.mean(losses[:-1])))
print("Combines Training/Testing Loss: {}".format(np.mean(losses)))
```

```python
def plot_compute_gaussian_scatter(labels, preds):
    cosmos = np.sort(np.unique(labels))    
    n = len(cosmos)
    preds_by_cosmo = np.zeros((n, 48, 1))
    for i in range(n):
        bool_cosmo = labels[:,0] == cosmos[i]
        preds_by_cosmo[i,:] = preds[bool_cosmo]
    preds_by_cosmo = np.reshape(preds_by_cosmo, (n, 48))
    mu_by_cosmo = np.mean(preds_by_cosmo, axis=1)
    std_by_cosmo = np.std(preds_by_cosmo, axis=1)
    cdf_by_cosmo = np.zeros(n)
    for i in range(n):
        cdf_by_cosmo[i] = st.norm.cdf(cosmos[i], mu_by_cosmo[i], std_by_cosmo[i])
    
    fig, axs = plt.subplots(nrows=5, ncols=4, figsize=(25, 20))
    fig.tight_layout()
    plt.subplots_adjust(left  = 0.125,  # the left side of the subplots of the figure
    right = 0.9,    # the right side of the subplots of the figure
    bottom = 0.1,   # the bottom of the subplots of the figure
    top = 0.9,      # the top of the subplots of the figure
    wspace = 0.5,   # the amount of width reserved for blank space between subplots
    hspace = 0.4,   # the amount of height reserved for white space between subplots)
    )
    for i in range(5):
        for j in range(4):
            ind = 4*i+j
            axs[i, j].plot(np.sort(preds_by_cosmo[ind]), 
                           st.norm.pdf(np.sort(preds_by_cosmo[ind]), mu_by_cosmo[ind], std_by_cosmo[ind]),
                           color='red', linewidth=8, label='normal fit')
            axs[i, j].hist(preds_by_cosmo[ind], bins=24, density=True, color='blue', label='model scatter')
            axs[i, j].axvline(x=cosmos[ind], c='g', linewidth=8, label='ground truth')
            axs[i, j].set_xlabel("Predictions for $\sigma_8={}$".format(round(cosmos[ind], 3)))
            axs[i, j].set_ylabel("Frequency")
            axs[i, j].set_title("Distribution of Predictions for $\sigma_8={}$".format(round(cosmos[ind], 3)))
            axs[i, j].legend()
    
    return (mu_by_cosmo, std_by_cosmo, cdf_by_cosmo)
```

```python
def plot_full_scatter(labels, preds):
    cosmos = np.sort(np.unique(labels))
    n = len(cosmos)
    preds_by_cosmo = np.zeros((n, 48, 1))
    for i in range(n):
        bool_cosmo = labels[:,0] == cosmos[i]
        preds_by_cosmo[i,:] = preds[bool_cosmo]
    preds_by_cosmo = np.reshape(preds_by_cosmo, (n, 48))
    std_by_cosmo = np.std(preds_by_cosmo, axis=1)
    linreg = np.polyfit(cosmos, std_by_cosmo, 1)
    p = np.poly1d(linreg)
    plt.plot(cosmos, std_by_cosmo, 'b-', label='scatter')
    plt.plot(cosmos, p(cosmos), 'r-', label='linear fit')
    plt.xlabel("Ground-truth Value of $\sigma_8$")
    plt.ylabel("Scatter in Model Predictions")
    plt.title("Model Scatter as a Function of $\sigma_8$")
    return linreg
```

```python
plot_full_scatter(np.vstack((y_train1, y_train2, y_train3, y_train4, y_test)), 
                  np.vstack((pred_train1, pred_train2, pred_train3, pred_train4, pred_test)))
```

```python
def plot_full_bias(labels, preds):
    cosmos = np.sort(np.unique(labels))
    n = len(cosmos)
    preds_by_cosmo = np.zeros((n, 48, 1))
    for i in range(n):
        bool_cosmo = labels[:,0] == cosmos[i]
        preds_by_cosmo[i,:] = preds[bool_cosmo]
    preds_by_cosmo = np.reshape(preds_by_cosmo, (n, 48))
    mu_by_cosmo = np.mean(preds_by_cosmo, axis=1)
    linreg = np.polyfit(cosmos, mu_by_cosmo-cosmos, 1)
    p = np.poly1d(linreg)
    plt.plot(cosmos, mu_by_cosmo-cosmos, 'b-', label='bias')
    plt.plot(cosmos, p(cosmos), 'r-', label='linear fit')
    plt.xlabel("Ground-truth Value of $\sigma_8$")
    plt.ylabel("Bias in Model Predictions")
    plt.title("Model Bias as a Function of $\sigma_8$")
    return linreg
```

```python
plot_full_bias(np.vstack((y_train1, y_train2, y_train3, y_train4, y_test)), 
               np.vstack((pred_train1, pred_train2, pred_train3, pred_train4, pred_test)))
```
