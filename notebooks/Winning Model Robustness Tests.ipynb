```python
from deep_dss.utils import *
from deep_dss.models import *
import numpy as np
import healpy as hp
import matplotlib.pyplot as plt
```

```python
model = model_by_architecture("winning")
```

```python
losses = []
for noise_scale in np.linspace(0.0425, 0.045, num=9):
    if noise_scale == -1:
        data = split_poisson_maps_by_dataset("TEST", multiplier=1.0, noiseless=True, 
                                             path_to_output="../data/flask/output/")
    else:
        data = split_poisson_maps_by_dataset("TEST", multiplier=1.0, density=noise_scale, 
                                             path_to_output="../data/flask/output/")
    preds = model.predict(data["x"])
    losses.append(l1_loss(preds, data["y"]))
```

```python
losses
```
