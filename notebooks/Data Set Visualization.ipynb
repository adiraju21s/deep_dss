```python
def plot_cl(ax, cl, lmax=3000, title=""):
        
        ax.set_title(title)
        ax.set_xlabel(r"Angular Scale, l")
        ax.set_ylabel(r"Power, C(l)")
        ax.set_yscale("log")
        ax.plot(cl["L"].to_numpy()[:lmax], cl["CL"].to_numpy()[:lmax], "b-")

def plot_all_cls(sigma8): 
    cls = load_cls(sigma8)
    
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
    
    plot_cl(axs[0,0], cls["gg"], title=r"Matter x Matter")
    plot_cl(axs[0,1], cls["gk"], title=r"Matter x Kappa")
    plot_cl(axs[1,0], cls["kk"], title=r"Kappa x Kappa")
    plot_cl(axs[1,1], cls["kk_"], title=r"Uncorrelated Kappa x Uncorrelated Kappa")
    fig.suptitle(r"Angular Power Spectra for $\sigma_8 = {}$".format(sigma8), fontsize=18)
    fig.show()
```

```python
def plot_psd(ax, cl, lmax=3000, title=""):
        
        ax.set_title(title)
        ax.set_xlabel(r"Angular Scale, l")
        ax.set_ylabel(r"Power Spectral Density, C(l)l(l+1)")
        ax.set_yscale("log")
        ax.plot(cl["L"].to_numpy()[:lmax], 
                cl["CL"].to_numpy()[:lmax]*cl["L"].to_numpy()[:lmax]*(cl["L"].to_numpy()[:lmax]+1), "b-")

def plot_all_psds(sigma8): 
    cls = load_cls(sigma8)
    
    fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
    
    plot_psd(axs[0,0], cls["gg"], title=r"Matter x Matter")
    plot_psd(axs[0,1], cls["gk"], title=r"Matter x Kappa")
    plot_psd(axs[1,0], cls["kk"], title=r"Kappa x Kappa")
    plot_psd(axs[1,1], cls["kk_"], title=r"Uncorrelated Kappa x Uncorrelated Kappa")
    fig.suptitle(r"Power Spectral Densities for $\sigma_8 = {}$".format(sigma8), fontsize=18)
    fig.show()
```
